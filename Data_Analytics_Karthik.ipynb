```python
salary = np.random.randint(200,900,size=5000)
salary
```

```python
sb.kdeplot(salary)
```

```python
# onesampletest
salary_mean = salary.mean()
salary_mean
```

```python
samplesize = 250
sal_sample = np.random.choice(salary,samplesize)
sal_sample
```

```python
from scipy import stats
```

```python
a,b,c,d=1,2,3,4
```

```python
a
```

```python
ttest,pvalue=stats.ttest_1samp(sal_sample,salary_mean)
ttest,pvalue
```

```python
if pvalue < 0.05:
    print('Null Hypothesis Rejected , HA Accepted, Independent')
else:
    print('Null Hypothesis Accepted , HA Rejected, Dependent')
```

```python
dep = 0
indep = 0
for i in range(1000):
    sal_sample = np.random.choice(salary,samplesize)
    ttest,pvalue=stats.ttest_1samp(sal_sample,salary_mean)
    if pvalue < 0.5:
        print('Null Hypothesis Rejected , HA Accepted, Independent------')
        print(ttest,pvalue)
        indep +=1
    else:
        print('Null Hypothesis Accepted , HA Rejected, Dependent')
        dep +=1
```

```python
dep,indep
```
